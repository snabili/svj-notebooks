```python
def build_feature_array(datasets, n=100):
    if hasattr(datasets, 'treename'): datasets = [datasets]
    to_flatten = []
    for dataset in datasets:
        arrays = next(dataset.iterate_batch(n))
        feature_array = svjflatanalysis.arrayutils.to_feature_array(arrays)
        to_flatten.append(feature_array)
    return np.concatenate(to_flatten, axis=-1).T
```

```python
def best_split(X, y):
    """Find the best split for a node.
    "Best" means that the average impurity of the two children, weighted by their
    population, is the smallest possible. Additionally it must be less than the
    impurity of the current node.
    To find the best split, we loop through all the features, and consider all the
    midpoints between adjacent training samples as possible thresholds. We compute
    the Gini impurity of the split generated by that particular feature/threshold
    pair, and return the pair with smallest impurity.
    Returns:
        best_idx: Index of the feature for best split, or None if no split is found.
        best_thr: Threshold to use for the split, or None if no split is found.

    Shamelessly copied/adapted from https://towardsdatascience.com/decision-tree-from-scratch-in-python-46e99dfea775
    """
    # Need at least two elements to split a node.
    m = y.size
    if m <= 1:
        return None, None

    # Count of each class in the current node.
    
    n_classes = 2
    num_parent = [np.sum(y == c) for c in range(n_classes)]

    # Gini of current node.
    best_gini = 1.0 - sum((n / m) ** 2 for n in num_parent)
    best_idx, best_thr = None, None

    
    n_features = X.shape[1]
    
    # Loop through all features.
    for idx in range(n_features):
        # Sort data along selected feature.
        thresholds, classes = zip(*sorted(zip(X[:, idx], y)))

        # We could actually split the node according to each feature/threshold pair
        # and count the resulting population for each class in the children, but
        # instead we compute them in an iterative fashion, making this for loop
        # linear rather than quadratic.
        num_left = [0] * n_classes
        num_right = num_parent.copy()
        for i in range(1, m):  # possible split positions
            c = classes[i - 1]
            num_left[c] += 1
            num_right[c] -= 1
            gini_left = 1.0 - sum(
                (num_left[x] / i) ** 2 for x in range(n_classes)
            )
            gini_right = 1.0 - sum(
                (num_right[x] / (m - i)) ** 2 for x in range(n_classes)
            )

            # The Gini impurity of a split is the weighted average of the Gini
            # impurity of the children.
            gini = (i * gini_left + (m - i) * gini_right) / m

            # The following condition is to make sure we don't try to split two
            # points with identical values for that feature, as it is impossible
            # (both have to end up on the same side of a split).
            if thresholds[i] == thresholds[i - 1]:
                continue

            if gini < best_gini:
                best_gini = gini
                best_idx = idx
                best_thr = (thresholds[i] + thresholds[i - 1]) / 2  # midpoint

    return best_idx, best_thr

best_split(X, y)
```

```python
X_cut = X
y_cut = y
for i in range(4):
    n_precut = y_cut.shape[0]
    i_col, cut_value = best_split(X_cut, y_cut)

    # Still need to determine which way the cut should point
    less_than_selection = X_cut[:,i_col] < cut_value
    greater_than_selection = ~less_than_selection

    y_cut_less_than = y_cut[less_than_selection]
    s_over_b_less_than = np.sum(y_cut_less_than) / y_cut_less_than.shape[0]
    y_cut_greater_than = y_cut[greater_than_selection]
    s_over_b_greater_than = np.sum(y_cut_greater_than) / y_cut_greater_than.shape[0]
    
    if s_over_b_less_than > s_over_b_greater_than:
        selection = less_than_selection
        y_cut = y_cut_less_than
        operator = '<'
    else:
        selection = greater_than_selection
        y_cut = y_cut_greater_than
        operator = '>'
            
    X_cut = X_cut[selection]
    
    n_postcut = y_cut.shape
    n_signal_postcut = np.sum(y_cut)
    
    print(
        'i_col={} cut_value: {}{} signal_fraction={}'
        .format(
            i_col, operator, cut_value, n_signal_postcut / n_postcut
            )
        )
```

```python
def sb(X, y):
    """
    Returns the sorted feature array, likewise-sorted label array, and the cumulative signal and background counts
    """
    # Sort the arrays
    sorter = X.argsort()
    X = X[sorter]
    y = y[sorter]
    s = np.cumsum(y==1)
    b = np.cumsum(y==0)
    s_reverse = np.cumsum(np.flip(y)==1)
    b_reverse = np.cumsum(np.flip(y)==0)
    return X, y, s, b, s_reverse, b_reverse
    
def ams(s, b, b_r=100):
    """ b_r: constant regularization term """
    return np.sqrt( 2 * ( (s + b + b_r) * np.log ( 1 + s / (b + b_r) ) - s ) )

def best_split_soversqrtb(X, y, do_ams=False):
    n_features = X.shape[1]
    best_i_feature = None
    best_feature_cut = None
    best_s_over_sqrtb = 0.
    for i_feature in range(n_features):
        sorted_x, sorted_y, s, b, s_reverse, b_reverse = sb(X[:,i_feature], y)
                
        if do_ams:
            metric = lambda s, b: ams(s, b)
        else:
            metric = lambda s, b: svjflatanalysis.arrayutils.safe_divide(s.astype(np.float), np.sqrt(b))
        
        s_over_sqrtb_lessthan = metric(s, b)
        s_over_sqrtb_greaterthan = metric(s_reverse, b_reverse)

        if np.max(s_over_sqrtb_lessthan) > np.max(s_over_sqrtb_greaterthan):
            s_over_sqrtb = s_over_sqrtb_lessthan
            operator = '<'
        else:
            s_over_sqrtb = s_over_sqrtb_greaterthan
            operator = '>'
            sorted_x = np.flip(sorted_x)
            sorted_y = np.flip(sorted_y)
                
        i_max = np.argmax(s_over_sqrtb)
        n_sig_left = np.sum(sorted_y[:i_max+1]==1)
        n_bkg_left = np.sum(sorted_y[:i_max+1]==0)
        print(
            'Best cut for feature {}: val {} {:.2f} (s/sqrt(b)={:.3f}, s={}, b={})'
            .format(
                svjflatanalysis.arrayutils.feature_labels[i_feature],
                operator, sorted_x[i_max], s_over_sqrtb[i_max], n_sig_left, n_bkg_left
                )
            )
        if s_over_sqrtb[i_max] > best_s_over_sqrtb:
            best_s_over_sqrtb = s_over_sqrtb[i_max]
            best_i_feature = i_feature
            best_feature_cut = sorted_x[i_max]
            
    return best_i_feature, best_feature_cut, best_s_over_sqrtb, operator

    
X_cut = X
y_cut = y
for i in range(4):
    n_precut = y_cut.shape[0]
    i_feature, cut, s_over_sqrtb, operator = best_split_soversqrtb(X_cut, y_cut, do_ams=True)

    if i_feature is None:
        print('No further good cut to make!')
        break
    
    print(
        'MAKING CUT: {} {} {} (metric={})'
        .format(
            svjflatanalysis.arrayutils.feature_labels[i_feature],
            operator, cut, s_over_sqrtb
            )
        )
    
    if operator == '<':
        selection = X_cut[:,i_feature] < cut
    else:
        selection = X_cut[:,i_feature] > cut
        
    X_cut = X_cut[selection]
    y_cut = y_cut[selection]
        
    
```
